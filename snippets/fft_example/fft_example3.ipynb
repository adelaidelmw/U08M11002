```python
# This is a Frequency analysis for a piece of music.
from pydub import AudioSegment
import matplotlib.pyplot as plt

# load mp3 file
audio = AudioSegment.from_file("test.mp3", format="mp3")
# audio = audio.set_frame_rate(44100).set_channels(1)

# convert to numpy array
samples = audio.get_array_of_samples()
# samples = samples.reshape(-1, audio.channels)

# plot audio waveform
plt.plot(samples[20000:20000+1000])
plt.show()
```

```python
# play audio
from pydub.playback import play

play(audio)
```

```python
print(samples[-1000:-900])
```

```python
import numpy as np
sample_rate = 44100  # unit: Hz

# calculate FFT of audio signal
fft = np.fft.fft(samples)

# calculate frequency bins
freq_bins = np.fft.fftshift(np.fft.fftfreq(len(samples), 1/sample_rate))

# plot spectrum
# plt.plot(freq_bins, np.abs(fft))
plt.plot(np.abs(fft))
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.show()
```
